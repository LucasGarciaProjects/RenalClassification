# Final Project Conclusion: Kidney Disease Image Classification

## Summary

This project aimed to develop a binary image classifier capable of distinguishing between healthy and diseased kidneys using pre-trained deep learning architectures. We explored three powerful convolutional neural network models: **MobileNetV2**, **ResNet50**, and **EfficientNetB0**, applying a range of transfer learning techniques and custom optimizations to maximize diagnostic performance.

## Challenges Encountered

Throughout the project, we faced significant obstacles that shaped our methodology and solution design:

- **Small and imbalanced dataset**: The dataset had a much larger number of diseased cases compared to healthy ones (230 vs. 68), which skewed early model predictions.
- **Initial bias towards one class**: All models initially predicted only one class (diseased), failing to detect healthy samples. This highlighted the need for threshold tuning and advanced fine-tuning techniques.
- **Overfitting risks**: Without careful control, validation loss often plateaued or worsened while training accuracy continued to rise.

To address these challenges, we applied:

- Progressive unfreezing of model layers
- Learning rate tuning and early stopping
- Custom threshold optimization using F1-score and balanced accuracy
- ROC curve analysis to understand trade-offs

## Future Work

If more time or data were available, we recommend the following:

- **Model Deployment**: Export ResNet50 and EfficientNetB0 models to TensorFlow Lite or ONNX for embedded or mobile deployment.
- **Data Augmentation**: Enhance training with class-specific augmentations to mitigate imbalance.
- **Explainability**: Integrate Grad-CAM or SHAP to visualize model decision-making.
- **Clinical Validation**: Test models on real-world, unseen renal images.
- **Ensemble Learning**: Combine the strengths of multiple architectures to improve reliability.

---

## Final Thoughts

This project demonstrates that with the right optimization strategies, even limited and imbalanced medical image datasets can yield useful, high-performing classifiers. While none of the models were perfect, ResNet50 and EfficientNetB0 in particular show strong potential for aiding kidney disease diagnosis in clinical settings.
